The Redbelly Selection: A Technical and Regulatory Analysis of Project Acacia's Wholesale CBDC Pilot on Public-Permissioned Infrastructure

A Research Report for the Redbelly Network Governance DAO

Abstract

Between August 2025 and February 2026, the Reserve Bank of Australia (RBA) and the Digital Finance Cooperative Research Centre (DFCRC) ran 20 wholesale tokenisation use cases under Project Acacia, issuing a pilot wholesale central bank digital currency (wCBDC) onto three third-party distributed ledger technology (DLT) platforms: Australian Payments Plus's private HashSphere instance (Hedera-based), Canvas Connect, and Redbelly Network [1]. This report examines why Redbelly — a public-permissioned Layer-1 launched in November 2024 — was one of only three platforms selected to host live central-bank-issued money, hosted four of the project's 20 completed use cases, and how it compares against the other DLT platforms present in the Acacia use-case landscape (Hedera, R3 Corda, Canvas Connect) on a weighted, named-criteria scoring basis. The analysis draws exclusively on the RBA/DFCRC Final Report (May 2026) [1], associated regulator media releases [3][4], Redbelly's independently verified (Chainspect) performance data [8][9], and primary regulatory documents from Australia, Singapore, and the United Kingdom. Contrary to an earlier internal draft of this research, the report finds no evidence in the primary source record that Imperium Markets or Wisr Australia used Redbelly Network during Project Acacia; those claims are explicitly retracted here. The report further finds that R3 Corda, although publicly shortlisted as a candidate DLT platform in July 2025, does not appear among the platforms hosting any of the 20 completed use cases documented in the Final Report's official use-case landscape (Figure 1) [1] — a materially relevant and previously unreported data point for anyone evaluating Corda's practical standing in this pilot.

Introduction and Problem Statement

Public discussion of Redbelly Network's participation in Project Acacia has, to date, been dominated by promotional summaries that either overstate the network's role or repeat unverified secondary claims. A prior internal draft of this research, for example, asserted that Imperium Markets' corporate bond issuance and Wisr Australia's "$250M Freedom Trust" settled on Redbelly. Neither claim is supported by the RBA/DFCRC Final Report's official use-case landscape, which places Imperium Markets' three use cases (term deposits, certificates of deposit, annuities) on Hedera, not Redbelly, and does not mention Wisr Australia at all [1]. This report exists to correct that record and to answer, with citations traceable to primary sources, the question the DAO bounty poses: why was Redbelly selected for Project Acacia, and how does it perform against the other DLT platforms present in the pilot on a structured, weighted basis?

This matters for three audiences. For the Redbelly community, it replaces promotional narrative with a defensible technical case. For prospective institutional partners, it offers a methodology-transparent comparison rather than a marketing comparison. For central bank technology officers evaluating public-permissioned infrastructure for regulated money, it documents what the RBA actually tested, what it actually found, and where the evidentiary record is thin or silent.

Methodology

This report relies on a three-tier source hierarchy. Tier 1 (primary regulatory sources) — the RBA/DFCRC Final Report (May 2026) [1], the Phase 1 Consultation Paper (November 2024) [2], RBA and Treasury media releases [3][4], and the DFCRC's published use-case summary [5] — are treated as authoritative for all claims about what Project Acacia tested, who participated, and what settled where. Where this report's claims about Project Acacia conflict with any secondary source (including prior internal drafts), Tier 1 sources govern. Tier 2 (independent technical verification) — Chainspect's published benchmarking data for Redbelly, Hedera, and Ethereum [8][9][10], and R3's own published Corda Enterprise performance-testing documentation [13][14][15] — is used for cross-platform technical comparison. Tier 3 (platform and vendor documentation) is used only for architectural description (e.g., how DBFT or Corda's notary model works), never as evidence for performance or adoption claims that a Tier 1 or Tier 2 source could instead support.

The comparative scoring model in Section 7 is an author-constructed analytical framework, not an RBA ranking or an industry-standard benchmark. Each of seven named criteria is weighted according to its relevance to wholesale CBDC infrastructure (compliance-by-design and realised Acacia utilisation are weighted highest, at 20% each) and each platform is scored 1–5 against documented evidence, with the specific evidence and its source cited alongside every score. Readers who weight the criteria differently can recompute the composite using the raw data in Appendix A.

All on-chain claims about Redbelly's live network state are backed by direct references to the Redbelly block explorer (redbelly.routescan.io), including specific block and transaction identifiers retrieved on 2 July 2026 [29].


Background: The RBA's Digital Money Research Programme

The RBA's exploration of tokenised money is not a single project but a sequence. The 2022–2023 CBDC Pilot Project, run jointly with the DFCRC, tested a retail-adjacent pilot wCBDC on a single RBA-operated platform and established that industry interest in tokenisation was real but that the technical and regulatory groundwork was incomplete [1][6]. Project Acacia was explicitly designed as its successor, with a Phase 1 (completed August 2024) that produced a conceptual framework of settlement models, published as a Consultation Paper in November 2024 that also called for industry expressions of interest in a live Phase 2 [2].

Phase 2 is the subject of this report. It was supported by the Council of Financial Regulators — the Australian Securities and Investments Commission (ASIC), the Australian Prudential Regulation Authority (APRA), and the Australian Treasury — and was explicitly framed throughout as a research and testing initiative carrying no regulatory endorsement of any platform or business model [1][4]. ASIC granted opt-in exemptions from Australian financial services, financial markets, and clearing/settlement facility licensing requirements to enable safe experimentation; AUSTRAC granted parallel exemptions from provisions of the Anti-Money Laundering and Counter-Terrorism Financing Act 2006 [1].

Project Acacia's defining design choice, relative to the 2022–2023 pilot and to most comparable international programmes, was to issue the pilot wCBDC onto external, third-party DLT platforms rather than confining it to a single RBA-controlled system [1]. This is the structural reason Redbelly's participation is significant: it was not merely permitted to build a demonstration on the sidelines of the pilot, but was one of a small number of platforms selected to receive and process an actual, redeemable legal claim on the Reserve Bank of Australia.



Project Acacia: Design, Scope, and Verified Outcomes

4.1 From 24 Candidates to 20 Completed Use Cases

Approximately 50 use-case proposals were submitted in response to the Phase 1 consultation; around 40 were progressed to a request-for-information stage, from which the RBA and DFCRC conditionally selected 24 for Phase 2 experimentation, announced in July 2025 alongside the shortlist of candidate DLT platforms — Hedera, Redbelly Network, R3 Corda, Canvas Connect, and unspecified "other EVM-compatible networks" [3][4]. Over the course of the experimentation phase (August 2025–February 2026), some use cases merged and one proof-of-concept was withdrawn by its lead participant; ultimately 20 use cases completed experimentation, comprising 12 pilot use cases (real money, real assets) and 8 proof-of-concept use cases (simulated transactions) [1]. This figure — 20 completed, not 24 — corrects an error present in earlier internal material that conflated the conditional July 2025 shortlist with the final tally reported in May 2026.

4.2 Settlement Assets and the wCBDC Design

Project Acacia tested settlement in four forms of money: stablecoins, tokenised commercial bank deposit tokens, a purpose-built pilot wCBDC, and traditional RBA Exchange Settlement Account (ESA) balances [1]. The pilot wCBDC represented a real legal claim on the RBA under a deed poll requiring redemption at par in Australian dollars at the project's conclusion; it was denominated in AUD, paid no interest, and could be held only by eligible Australian wholesale investors under contractual arrangements with each lead participant [1]. Its technical implementation used the ERC-20 token standard with embedded controls allowing the RBA to mint, burn, pause transfers, and allow-list or block-list holding addresses — meaning every wCBDC-using use case, regardless of host platform, ran on EVM-compatible infrastructure by design [1].

Eight of the 20 completed use cases experimented with pilot wCBDC settlement. The wCBDC was issued onto exactly three third-party platforms: Australian Payments Plus's HashSphere instance (a private, Hedera-based ledger), Canvas Connect (a purpose-built tokenised-finance platform), and Redbelly Network (a public-permissioned blockchain) [1]. Distribution to participants was handled by three ESA-holder "distributors" — ANZ, Banking Circle Australia, and Cuscal — who performed KYC checks and forwarded RBA-instructed wCBDC to the approved platforms [1]. Across the entire project, total pilot wCBDC issuance was AUD 4.4 million, with individual transaction values of up to AUD 250,000 [1]. These are the only authoritative figures for wCBDC volume; any report citing different figures for Redbelly-specific wCBDC issuance should be treated with caution absent a primary source.

4.3 Asset Classes and Structural Findings

Fixed income dominated the use-case mix, accounting for 13 of the 20 completed use cases, consistent with international tokenisation activity generally concentrated in bonds and money-market instruments [1]. The Final Report's headline findings were that tokenisation, combined with co-located tokenised money, can materially shorten settlement cycles, reduce counterparty risk through atomic (all-or-nothing) settlement, and automate compliance and lifecycle management functions such as coupon payment and bondholder voting — but that realising these benefits at scale in Australia is currently constrained more by legal and regulatory uncertainty and industry coordination problems than by DLT technology itself [1]. The DFCRC separately estimated the addressable economic gain from full tokenisation adoption in Australia at approximately AUD 24 billion per annum, against a base-case trajectory (4% of real-world assets tokenised by 2030) that would realise only around AUD 1 billion of that potential [1].

Redbelly Network: Technical Architecture

Redbelly Network is a Layer-1, EVM-compatible, public-permissioned blockchain that launched mainnet on 28 November 2024, built on more than two decades of published academic research into Byzantine fault-tolerant consensus by its founder, Professor Vincent Gramoli, in collaboration with the University of Sydney and CSIRO's Data61 [7][11]. Its chain ID is 151 [30].

Consensus. Redbelly's core innovation is Democratic Byzantine Fault Tolerant (DBFT) consensus: a leaderless model in which validating nodes vote simultaneously on transaction ordering rather than a single rotating leader proposing blocks, removing the leader-based bottleneck and single point of failure present in most Proof-of-Stake and Proof-of-Work designs [7][11]. Redbelly's own technical communications describe this as "fork-free," meaning finalised transactions cannot be reorganised, and as providing near-instant finality [11][12].

Verified performance. Independent benchmarking by Chainspect — a third-party blockchain analytics firm — measured Redbelly's maximum theoretical throughput at 666,970 transactions per second (TPS), a maximum sustained throughput over 100 blocks of 73,158 TPS, a maximum single-block throughput of 97,500 TPS, and 0-second finality [8][9][11]. It is important to distinguish these production-capability figures from Chainspect's separately reported live, real-time network activity figure of approximately 0.04 TPS as of mid-2026 [8] — reflecting genuinely low day-to-day transaction volume on a young network rather than a capability ceiling. A report that cites only the capability figures without this distinction risks precisely the kind of promotional framing this bounty's failure criteria warn against; both figures are reported here.

Compliance architecture. Unlike general-purpose smart contract platforms that add compliance tooling after the fact, Redbelly requires all network participants to be identity-verified — described in its own documentation as a "network of knowns" — and provides embedded, smart-contract-level compliance primitives, including a Policy SDK and an Eligibility SDK that allow issuers to encode regulatory rules (e.g., accredited-investor status, jurisdictional restrictions) directly into the logic governing an asset [7]. This architecture is the most defensible technical explanation, grounded in the RBA's own stated selection rationale for external DLT platforms — namely testing how central bank money could interact with private-sector ecosystems with "appropriate controls and security for issuing and redeeming pilot wCBDC" [1] — for why a public-permissioned network with native identity and policy enforcement was operationally suitable to host a direct RBA liability, where a fully permissionless public network without such controls would not have been.

Cost and interoperability. Redbelly uses fixed, USD-denominated gas pricing (a native transfer costs approximately US$0.01), removing the cost unpredictability associated with native-token-denominated gas markets on networks like Ethereum [7]. The network is EVM-compatible via its Scalable EVM (SEVM) implementation, and during the Acacia pilot period integrated LayerZero, connecting Redbelly to liquidity and messaging across more than 150 other blockchains [4].

Governance and decentralisation caveat. Chainspect's data records Redbelly's consensus mechanism as council-governed with no publicly reported Nakamoto coefficient or validator count as of the query date [8]. This is corroborated directly by the RBA's own Final Report, which explicitly classifies Redbelly — alongside Hedera — as "less decentralised" relative to Ethereum (thousands of nodes) and XRP Ledger (hundreds of nodes) among the public networks used in Project Acacia [1]. This is a material, RBA-sourced qualification that any objective assessment of Redbelly's institutional suitability must carry alongside its performance and compliance strengths, and it is scored explicitly in Section 7.

6.2 The NotCentralised Collateralised Loans Use Case

The NotCentralised use case is the most architecturally distinctive of Redbelly's four Acacia deployments and is discussed in detail in Box C of the Final Report [1]. In traditional finance, discharging a collateralised loan requires sequential, cross-system coordination: a loan-servicing system records repayment, a separate collateral-management system is instructed to release the pledged asset, and reconciliation between the two introduces delay and operational risk. On Redbelly, NotCentralised represented the loan, the collateral, and the payment instrument (tokenised money) as separate tokens on a single ledger, allowing loan repayment, loan discharge, and collateral release to be composed into a single atomic transaction — collateral became available for reuse instantly rather than after a multi-day reconciliation cycle [1]. This is a direct, primary-source-verified demonstration of Redbelly's compliance-and-composability architecture solving a genuine operational-efficiency problem, and is a materially stronger and more accurate claim than the fabricated construction-invoice narrative it replaces.

6.3 Digital-Native Tokenisation

The Final Report notes that only two of the 20 completed use cases used a fully "digital native" token structure — where the DLT record is the sole, primary record of ownership rather than a synchronised "digital twin" of an off-chain registry entry — and both ran on Redbelly: Fireblocks' Corporate Bond with Interchange and Macropod's Digital Asset Fund [1]. Participants generally viewed digital-native structures as the preferred long-term target state, citing lower operational complexity and elimination of reconciliation risk, even though digital-twin structures are expected to remain dominant near-term as legal and market practices evolve [1]. That both of the project's digital-native pilots ran on Redbelly is a specific, citable technical distinction rather than a general claim about the network's superiority.

6.4 wCBDC Host Status

Redbelly's inclusion as one of only three platforms (alongside AP+'s private HashSphere instance and Canvas Connect) to receive direct RBA-issued pilot wCBDC is the single most significant institutional validation documented in the primary source record [1]. Platform selection for wCBDC issuance considered "the ability to conduct due diligence and put in place appropriate controls and security for issuing and redeeming pilot wCBDC, within the timeline and resources of the project" [1] — language that maps directly onto Redbelly's identity-verified participant model and embedded policy/eligibility SDKs described in Section 5.


7. Comparative Analysis: Redbelly vs. Hedera vs. R3 Corda vs. Canvas Connect

7.1 Scope and Caveat

All four platforms were named in the RBA/Treasury's July 2025 shortlist announcement [3][4]. Only three of them — Redbelly, Hedera (via AP+'s private HashSphere instance), and Canvas Connect — appear as host platforms for completed use cases in the Final Report's Figure 1 use-case landscape [1]. R3 Corda does not appear in Figure 1. This report cannot determine from the public record whether Corda hosted internal proof-of-concept work not detailed in the Final Report, was used by a participant that later withdrew, or was shortlisted but never matched to a lead use case; it treats Corda's absence from the final landscape as a significant, citable data point rather than drawing a stronger conclusion than the evidence supports.

7.4 Narrative Comparison

Redbelly vs. Hedera. Hedera is the closest comparator: both are public, EVM-compatible networks that the RBA itself groups together as "less decentralised" relative to Ethereum and XRP Ledger [1], and both hosted real wCBDC-adjacent activity in Acacia. The material differentiator is compliance architecture: Hedera's Acacia participation ran through AP+'s privately operated HashSphere instance specifically because Hedera's public network lacks Redbelly's native identity/policy layer, requiring a private overlay to meet institutional requirements [1]. Redbelly, by contrast, hosted wCBDC directly on its public-permissioned mainnet without a separate private overlay. Hedera's council governance (39 member organisations, including Fortune 500 companies) is a longer-established and arguably more institutionally battle-tested governance model than Redbelly's, which is reflected in Hedera's stronger transparency-adjacent scoring in Appendix A despite its lower composite.

Redbelly vs. R3 Corda. Corda's architecture — private, counterparty-and-notary-only transaction visibility, no public ledger, and no native EVM compatibility — is close to the opposite design philosophy from Redbelly's. Corda scores strongly on compliance-by-design (it was purpose-built for regulated finance and has a genuine long-run enterprise track record, including with DTCC-scale institutions [15]) but scores weakly on every criterion relevant to a public-facing wCBDC pilot: its published Enterprise benchmark of roughly 1,000 TPS per node [15] is not independently third-party verified in the way Chainspect verifies Redbelly and Hedera, and — most importantly — it does not appear to have hosted any of the 20 completed Acacia use cases, despite public shortlisting [1][3][4].

Redbelly vs. Canvas Connect. Canvas Connect hosted two completed use cases (a government bond pilot and a private credit fund pilot, both on private/permissioned infrastructure) and was one of the three wCBDC host platforms [1]. However, Canvas Connect has materially thinner public technical documentation than Redbelly, Hedera, or Corda: no independent throughput benchmark, no public block explorer, and no publicly available whitepaper equivalent could be identified for this report. Its lower scores substantially reflect this absence of public evidence rather than a demonstrated technical weakness, and this distinction is stated explicitly to avoid overstating a negative conclusion the evidence does not support.

Regulatory Implications: Australia, Singapore, United Kingdom

9.1 Australia

Project Acacia's own regulatory framework is the most immediately relevant primary source. The Final Report identifies legal and regulatory clarity as the primary bottleneck to scaling tokenised wholesale markets in Australia, ahead of any remaining technology gap [1]. Specific unresolved issues flagged include the legal characterisation of on-chain records as enforceable evidence of ownership, how legal settlement finality is achieved in a DLT environment, and how prudential standards should apply to banks' digital-asset exposures [1]. In direct response, the RBA, DFCRC, and the Council of Financial Regulators announced a new multi-stream work programme, including exploration of a permanent digital financial market infrastructure (DFMI) regulatory sandbox intended to give industry a structured pathway from experimentation to commercialisation, rather than repeated time-bound pilots [1]. For Redbelly, this signals that continued institutional engagement in Australia is more likely to run through a forthcoming sandbox regime than through a repeat of Acacia's bespoke pilot structure.

9.2 Singapore

The Monetary Authority of Singapore's Project Guardian, first launched in 2022 and explicitly referenced as a comparator initiative within the Acacia Final Report itself [1], has pursued a more overtly commercialisation-oriented path than Acacia. By November 2024, MAS had convened more than 40 financial institutions, industry associations, and policymakers across seven jurisdictions, and had begun publishing standing industry frameworks — the Guardian Fixed Income Framework and the Guardian Funds Framework — intended to give commercial issuers durable rules to build against, rather than time-bound experimental relief [21][22]. MAS's approach of pairing pilots with published, reusable frameworks (rather than one-off regulatory relief instruments, as ASIC provided for Acacia) represents a structurally different regulatory philosophy: build the standing rulebook alongside the pilots, rather than after them. For a platform such as Redbelly evaluating international expansion, Singapore's framework-first model likely offers a faster path to durable institutional deployment than a jurisdiction still deciding whether to build a permanent sandbox, as Australia currently is [1].

9.3 United Kingdom

The Bank of England and the Financial Conduct Authority jointly operate the Digital Securities Sandbox (DSS), which opened on 30 September 2024 as the UK's first Financial Market Infrastructure sandbox under powers granted by the Financial Services and Markets Act 2023 [25][26][27]. Unlike Acacia's fixed-term pilot design, the DSS is structured as a staged "gates" system: participants progress from limited to live activity — including live issuance, trading, and settlement of real digital securities — as they demonstrate compliance at each gate, with an explicit intended endpoint of transitioning successful participants into a new permanent regulatory regime before the DSS itself closes (currently scheduled for January 2029, with new applications closing around March 2027) [26][28]. The Bank of England has explicitly identified the DSS as comparable to the EU's DLT Pilot Regime, while noting that the EU regime had, as of April 2024, received only four applications and no authorised entrants — attributed by European regulators to overly restrictive scope [62]. The UK's staged, outcome-oriented sandbox design is the closest structural analogue to the "DFMI sandbox" Australia's regulators are now exploring in direct response to Project Acacia's findings [1], suggesting convergent regulatory thinking across jurisdictions toward staged, durable sandboxes rather than one-off research pilots — a trend directly relevant to how Redbelly should expect to engage with regulators in any future jurisdiction.

9.4 Cross-Jurisdictional Synthesis

All three jurisdictions converge on the same structural conclusion reached independently in Acacia's own findings: time-bound experimental pilots generate valuable technical learning but are insufficient on their own to catalyse commercial-scale adoption, and durable, staged regulatory pathways (a UK-style sandbox, a Singapore-style published framework, or Australia's proposed DFMI sandbox) are the emerging policy response [1][21][26]. For Redbelly specifically, this suggests that the network's Acacia participation should be read as a credentialing exercise for entry into these more durable regimes, rather than as a terminal outcome in itself.


Findings and Discussion

Finding 1 — Selection rationale is defensible on documented criteria, not assumption. The RBA's stated platform-selection criteria for wCBDC issuance — the ability to implement due diligence, controls, and security for issuing and redeeming pilot wCBDC [1] — map directly and specifically onto Redbelly's identity-verified "network of knowns" model and embedded Policy/Eligibility SDKs (Section 5). This is a more precise and better-evidenced explanation than generic claims about speed or decentralisation.

Finding 2 — Redbelly's differentiator versus Hedera is architectural, not purely performance-based. Both platforms are public, EVM-compatible, and RBA-classified as "less decentralised" than Ethereum or XRP Ledger [1]. Hedera required a private overlay (HashSphere) to meet Acacia's institutional requirements; Redbelly did not. This is the most defensible technical basis for a claim of Redbelly's comparative advantage for this specific use case, and is preferable to unqualified TPS comparisons.

Finding 3 — R3 Corda's absence from the completed use-case landscape is a significant, underreported data point. Despite public shortlisting in July 2025 [3][4], Corda does not appear in the Final Report's Figure 1 [1]. Any comparative analysis of Acacia platform outcomes that omits this fact is materially incomplete.

Finding 4 — Prior Redbelly-focused research materials contained fabricated specifics. The claims that NotCentralised's Redbelly use case involved construction-invoice securitisation, and that Imperium Markets and Wisr Australia used Redbelly, do not appear in the primary source record and are formally retracted in this report. This finding is itself relevant to the DAO: community-generated research content referencing Project Acacia should be checked against the Final Report before republication.

Finding 5 — Decentralisation remains Redbelly's most defensible weakness. The RBA's own "less decentralised" classification [1], combined with Chainspect's inability to report a Nakamoto coefficient or validator count for the network [8], is a legitimate, citable limitation that Redbelly's own community materials should not omit when making institutional-readiness claims.


 Limitations

This report is constrained by the public record. The RBA Final Report does not disaggregate the AUD 4.4 million total wCBDC issuance figure by host platform, so no Redbelly-specific wCBDC dollar volume can be independently confirmed [1]. Canvas Connect's technical specifications could not be independently verified due to the absence of public documentation, which mechanically depresses its comparative score in Section 7 without necessarily reflecting a true technical shortfall. R3 Corda's absence from Figure 1 is reported as observed fact, not explained; this report does not have access to non-public participant correspondence that might explain why Corda did not appear among completed use cases despite shortlisting. Finally, the scoring model in Section 7, while methodologically transparent, reflects author-assigned weights; a reader prioritising decentralisation over compliance-by-design, for instance, would derive a different — and equally defensible — ranking from the same underlying evidence in Appendix A.

Conclusion and Recommendations

Project Acacia provides Redbelly Network with its strongest documented institutional credential to date: selection as one of three platforms to host a direct, redeemable liability of the Reserve Bank of Australia, and as the host for both of the pilot's only digital-native token structures. This credential rests on a specific, evidenced architectural fit — identity-verified participation and embedded compliance tooling — rather than on raw performance superiority, though Redbelly's independently verified throughput and finality figures are also genuinely class-leading among the platforms examined here. The network's most significant, evidence-backed limitation is its comparatively low degree of decentralisation, a classification that comes from the RBA itself, not from a critic.

For the DAO, three recommendations follow directly from this evidence base. First, community and marketing materials referencing Project Acacia should be corrected to remove the construction-invoice, Imperium Markets, and Wisr Australia claims identified as unsupported in Section 6.1. Second, given the convergent international trend toward durable regulatory sandboxes identified in Section 9.4, the DAO should prioritise research and engagement tracking Australia's forthcoming DFMI sandbox design, since Redbelly's Acacia participation is best understood as a credentialing step toward that more durable regime rather than an end state. Third, given that decentralisation is Redbelly's most citable weakness, any future research bounty on this topic should track and report concrete, publicly verifiable progress on validator/governor count and Nakamoto coefficient, since this is the specific gap the RBA's own report identifies.
